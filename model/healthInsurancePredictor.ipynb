```python
mainDF = pd.read_sql_query('SELECT * FROM healthData', conn)
```

```python
mainDF.head()
```

```python
features = mainDF.drop('charges', axis=1)
targets = mainDF['charges']
```

```python
from sklearn.model_selection import train_test_split

features_train, features_test, targets_train, targets_test = train_test_split(features, targets, test_size=.2, random_state=42)
```

```python
# Try linear regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

lr = LinearRegression()
lr.fit(features_train, targets_train)
chargesPredictionlr = lr.predict(features_test)
```

```python
print(r2_score(targets_test, chargesPredictionlr))
print(mean_squared_error(targets_test, chargesPredictionlr))
print(mean_absolute_error(targets_test, chargesPredictionlr))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators = 100, random_state=42)
rf.fit(features_train, targets_train)
chargesPredictionrf = rf.predict(features_test)
```

```python
print(r2_score(targets_test, chargesPredictionrf))
print(mean_squared_error(targets_test, chargesPredictionrf))
print(mean_absolute_error(targets_test, chargesPredictionrf))
```

```python
mainDF['region'] = mainDF['region'].map({0: 'northeast', 1: 'northwest', 2: 'southeast', 3: 'southwest'})
```

```python
# Use dummy variables instead of the categories for region
mainDFd = pd.get_dummies(mainDF, columns = ['region'], drop_first=True)
mainDFd.head()
```

```python
# Categorize bmi into categories and then use dummy variables
def bmiC(bmi):
    if bmi < 18.5:
        return 'underweight'
    elif bmi < 25:
        return 'normal'
    elif bmi < 30:
        return 'overweight'
    else:
        return 'obese'

mainDFd['bmi'] = mainDFd['bmi'].apply(bmiC)
mainDFd = pd.get_dummies(mainDFd, columns = ['bmi'], drop_first='True')
mainDFd.head()
```

```python
# Train new linear regression model
featuresd = mainDFd.drop('charges', axis=1)
targetsd = mainDFd['charges']
features_train2, features_test2, targets_train2, targets_test2 = train_test_split(featuresd, targetsd, test_size=.2, random_state=42)
lr2 = LinearRegression()
lr2.fit(features_train2, targets_train2)
chargesPredictionlr2 = lr2.predict(features_test2)
```

```python
print(r2_score(targets_test2, chargesPredictionlr2))
print(mean_squared_error(targets_test2, chargesPredictionlr2))
print(mean_absolute_error(targets_test2, chargesPredictionlr2))
```

```python
rf2 = RandomForestRegressor(n_estimators = 100, random_state=42)
rf2.fit(features_train2, targets_train2)
chargesPredictionrf2 = rf2.predict(features_test2)
```

```python
print(r2_score(targets_test2, chargesPredictionrf2))
print(mean_squared_error(targets_test2, chargesPredictionrf2))
print(mean_absolute_error(targets_test2, chargesPredictionrf2))
```

```python
# Plot models compared to actual
import matplotlib.pyplot as plt
# lr models compared to actual
plt.figure(figsize=(10,6))
plt.scatter(targets_test, chargesPredictionlr, label="LR Model 1", color='blue')
plt.scatter(targets_test, chargesPredictionlr2, label="LR Model 2", color='yellow')
plt.plot([targets_test.min(), targets_test.max()], [targets_test.min(), targets_test.max()], 'r', label='Perfect Accuaracy')
plt.xlabel('Actual')
plt.ylabel('Prediction')
plt.title('Linear Regression Models Accuracy')
plt.show()
```

```python
# RF models compared to actual
plt.figure(figsize=(10,6))
plt.scatter(targets_test, chargesPredictionrf, label="RF Model 1", color='blue')
plt.scatter(targets_test, chargesPredictionrf2, label="RF Model 2", color='yellow')
plt.plot([targets_test.min(), targets_test.max()], [targets_test.min(), targets_test.max()], 'r', label='Perfect Accuaracy')
plt.xlabel('Actual')
plt.ylabel('Prediction')
plt.title('Random Forest Models Accuracy')
plt.show()
```
